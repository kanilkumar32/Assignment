X and y Split

```python
df.head()
```

```python
X=df.loc[:,['curbweight','carwidth','carlength','enginesize','drivewheel','boreratio','fuelsystem','horsepower','wheelbase']]
X.head()
```

```python
y=df['price']
y.head()
```

Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scale =MinMaxScaler()
```

```python
scaled_x = pd.DataFrame(scale.fit_transform(X),columns = X.columns)
scaled_x.head()
```

Perform Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(scaled_x,y,test_size =0.2,random_state = 0)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train.head()
```

```python
x_test.head()
```

Train the data with suitable ML model

```python
from sklearn.linear_model import LinearRegression
lr =LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

Test the model

```python
x_test
```

```python
y_test
```

```python
price=pd.DataFrame({'Actual Price':y_test,'Predicted Price':y_pred})
price
```

Test with random values

```python
random_data = np.array([0.5,0.7,0.3,0.8,0.9,0.6,0.5,0.4,0.7])
lr.predict([random_data])
```

Evaluate the model

```python
from sklearn import metrics
```

```python
# R-Square
metrics.r2_score(y_test,y_pred)
```

```python
# MSE
metrics.mean_squared_error(y_test,y_pred)
```

```python
# RMSE

np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```
